In this project, I am exploring various earnings of graduated college students now in their respective career fields. This data, 'recent-grads.csv', was provided by the American Community Survey through surveys and aggregated data. FiveThirtyEight graciously cleaned the dataset. I will be using pandas to transform this data into easy to comprehend visualizations like scatter plots, histograms, and bar charts. From these, valuable insights can be gained more easily.

```python
import pandas as pd
import matplotlib as mpl

'''The output of plotting commands
is displayed inline within frontends directly below 
the code cell that produced it. 
The resulting plots will then also be stored in the 
notebook document.'''
%matplotlib inline
```

```python
#I will read the dataset into a DataFrame to explore it.
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

This tells alot. The major, gender distribution, employment type, and general job statistics (Median salary, top and bottom 25% of salaries) and sample size are shown for each degree.

```python
recent_grads.head()
#Getting the first and the last bits of the set to see what
#else it says about other majors.
```

```python
recent_grads.tail()
```

```python
#Getting summary statistics of all the numeric columns.
recent_grads.describe()
```

```python
#Just making sure that there are 173 different degrees I
#have information on.
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
```

```python
#Now dropping any rows that have missing values.
recent_grads = recent_grads.dropna(how = 'any')
```

```python
#Checking the rows again now that its cleaned.
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

Now we have a clean data set that has no empty values in it.

With cleaned data, we are now going to make some scatter plots to see if we can divine any relationships between the values.

```python
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter', title = 'Sample_size vs Median')
```

```python
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter', title = 'Sample_size vs Unemployment_rate')
```

```python
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter', title = 'Full Time vs Median')
```

```python
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter', title = 'Share Women vs Unemployment_rate')
```

```python
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter', title = 'Men vs Median')
```

```python
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter', title = 'Women vs Median')
```

Looking at the 'Unemployment_rate' vs. 'ShareWomen' plot, there seems to be no correlation between unemployment rate and the amount of women in the major.

Now scatter matrices will be employed to see if they can show any other information.

```python
from pandas.plotting import scatter_matrix
```

```python
#First Sample Size and Median will be plotted to see any possible correlation.
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (10,10))
```

```python
#Now Sample Size, Median, and Unemployment Rate will be used.
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

In all honesty, this does not help much, there is some negative correlation between SharWomen and Median. This could suggest that degrees that typically have less women earn more. 

Bar charts will be used last to see if anything else can be gleaned.

```python
recent_grads[:10].plot.bar(x='Major', y ='ShareWomen')
```

```python
recent_grads[-10:].plot.bar(x='Major', y ='ShareWomen')
```

Using the first ten degrees and the last ten, a disparity in the gender distribution can be seen with more men in the engineering related majors.
Maybe a salary gap can be seen here.

```python
recent_grads[:10].plot.bar(x='Major', y ='Unemployment_rate')
```

Using the first ten degrees and the last ten, a disparity in the gender distribution can be seen with more men in the engineering related majors.
Maybe a salary gap can be seen here.

```python
recent_grads[:10].plot.bar(x='Major', y ='Median')
```

```python
recent_grads[-10:].plot.bar(x='Major', y ='Median')
```

This assumption seems to be validated. For the selected majors, the ones with lower ShareWomen values tend to earn more.
